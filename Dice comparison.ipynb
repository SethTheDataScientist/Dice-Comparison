## Set up Functions

```python
def roll(die, verbose):
    result = random.choice(die)
    if verbose:
        print(result)
    return result
```

```python
def compare_dice(die1, die2, verbose):
    result = np.where(die1 >= die2, "Higher", "Lower")
    if verbose:
        print(result)
    return result
```

```python
def max_die_face(die,die_map, verbose):
    first_roll = roll(die, False)
    values = die_map[first_roll]
    highest_face = np.max(values)
    result = np.max(np.where(first_roll >= highest_face, first_roll, highest_face))
    if verbose:
        print(first_roll)
        print(values)
        print(highest_face)
        print(result)
    return result
```

```python
def minest_die_face(die, die_map, verbose):
    first_roll = roll(die, False)
    values = die_map[first_roll]
    highest_face = np.min(values)
    result = np.min(np.where(first_roll <= highest_face, first_roll, highest_face))
    # result = np.min(np.where(first_roll == 20, 20, result))
    if verbose:
        print(first_roll)
        print(values)
        print(highest_face)
        print(result)
    return result
```

```python
def min_die_face(die, die_map, verbose):
    first_roll = roll(die, False)
    values = die_map[first_roll]
    highest_face = np.min(values)
    result = np.min(np.where(first_roll >= highest_face, first_roll, highest_face))
    if verbose:
        print(first_roll)
        print(values)
        print(highest_face)
        print(result)
    return result
```

```python
def median_die_face(die,die_map, verbose):
    first_roll = roll(die, False)
    values = die_map[first_roll]
    middle_index = len(values) // 2
    highest_face = values[middle_index]
    result = np.max(np.where(first_roll >= highest_face, first_roll, highest_face))
    if verbose:
        print(first_roll)
        print(values)
        print(highest_face)
        print(result)
    return result
```

```python
def random_die_face(die,die_map, verbose):
    first_roll = roll(die, False)
    values = die_map[first_roll]
    highest_face = random.choice(values)
    result = np.max(np.where(first_roll >= highest_face, first_roll, highest_face))
    if verbose:
        print(first_roll)
        print(values)
        print(highest_face)
        print(result)
    return result
```

```python
def randest_die_face(die, die_map, verbose):
    first_roll = roll(die, False)
    values = die_map[first_roll]
    values.append(first_roll)
    result = random.choice(values)
    if verbose:
        print(first_roll)
        print(values)
        print(result)
    return result
```

```python
def roll_with_emphasis(die, verbose):
    value_1 = roll(die, False)
    value_2 = roll(die, False)
    roll_1 = abs(value_1 - 10.5)
    roll_2 = abs(value_2 - 10.5)
    if roll_1 > roll_2:
        result = value_1
    elif roll_1 < roll_2:
        result = value_2
    else:
        result = roll_with_emphasis(d20, False)
    
    if verbose:
        print(value_1)
        print(value_2)
        print(roll_1)
        print(roll_2)
        print(result)
    return result
```

```python
def advantage_result(die1, die2, advantage, verbose):
    if advantage:
        result = max(die1, die2)
    else:
        result = min(die1, die2)
    if verbose:
        print(result)
    return result
```

```python
def multi_advantage_result(dice, count, advantage, verbose):
    multi__list = []
    for i in range(count):
        multi__list.append(roll(dice, False))
    if advantage:
        result = np.max(multi__list)
    else:
        result = np.min(multi__list)
    if verbose:
        print(result)
    return result
```

```python
def average_result(dice, count, verbose):
    average_list = []
    for i in range(count):
        average_list.append(roll(dice, False))
    result = np.mean(average_list)
    if verbose:
        print(result)
    return result
```

## Random Checks

```python
def clear_number(number, die):
    value = roll(die, False)
    result = np.where(value >= number, 1, 0)
    return result
```

```python
clear_rolls = []
for i in range(100000):
    clear_rolls.append(clear_number(13, d20))

output = np.mean(clear_rolls)
print(output)
```

```python
def clear_multiple_dice(number, die, count):
    multi__list = []
    for i in range(count):
        multi__list.append(roll(die, False))
    min_value = np.min(multi__list)
    result = np.where(min_value >= number, 1, 0)
    return result
```

```python
clear_multiple = []
for i in range(100000):
    clear_multiple.append(clear_multiple_dice(8, d20, 2))

output = np.mean(clear_multiple)
print(output)
```

```python
list = [i+1 for i in range(20)]
print(list)
output_list =[]
for i in list:
    clear_rolls = []
    for j in range(100000):
        clear_rolls.append(clear_number(i, d20))

    output = np.mean(clear_rolls)
    output_list.append(output)

print(output_list)
```

```python
double_output_list =[]
for i in list:
    clear_multiple = []
    for j in range(100000):
        clear_multiple.append(clear_multiple_dice(i, d20, 2))

    output = np.mean(clear_multiple)
    double_output_list.append(output)

print(double_output_list)
```

```python
triple_output_list =[]
for i in list:
    clear_multiple = []
    for j in range(100000):
        clear_multiple.append(clear_multiple_dice(i, d20, 3))

    output = np.mean(clear_multiple)
    triple_output_list.append(output)

print(triple_output_list)
```

```python
output_table = {
    'single': output_list,
    "double": double_output_list,
    "triple": triple_output_list
}
```

```python
output_df = pd.DataFrame(output_table)
```
